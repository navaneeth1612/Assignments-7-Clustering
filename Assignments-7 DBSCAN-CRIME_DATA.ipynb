# Import libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

# Load Data

```python
crimeDB=crime.rename({'Unnamed: 0':'State'},axis=1)
crimeDB
```

```python
crimeDB.info()
```

```python
#dropping state column as it is not required here
crimeDB.drop(['State'],axis=1, inplace=True)
```

```python
crimeDB.head()
```

```python
#DBSCAN expects data in the form of an array
array=crimeDB.values
array
```

```python
#standardize the data
stdscaler=StandardScaler().fit(array)
x=stdscaler.transform(array)
```

```python
x
```

```python
#Parameters Required For DBSCAN Algorithm
#1.eps#One way to find the eps value is based on the k-distance graph.
#2.MinPts#the minimum MinPts can be derived from the number of dimensions D in the dataset as, MinPts >= D+1. The minimum value of MinPts must be chosen at least 3.
```

```python
dbscan = DBSCAN(eps=1.2, min_samples=5)
dbscan.fit(x)
```

```python
dbscan.labels_#-1 is noisy sample
```

```python
clust=pd.DataFrame(dbscan.labels_,columns=['cluster'])
clust.head()
```

```python
crimeDB1=pd.concat([crimeDB,clust],axis=1)
```

```python
crimeDB1.head()
```

```python
crimeDB1.groupby('cluster').agg(['mean']).reset_index()
```

```python
crimeDB1['cluster'].nunique()
```

```python
print(crimeDB1['cluster'].unique())
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(crimeDB1['cluster'],crimeDB1['Murder'], s=10, c=dbscan.labels_)
```
