# Import libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as sch
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

# Load Data

```python
crime=pd.read_csv(r'C:\Users\BIS\Desktop\DATA SCIENCE ASSIGNMENTS\Clustering\crime_data.csv')
```

```python
crime.head()
```

```python
crime.shape
```

```python
#Renaming the Unnamed: 0 Column into State
crime2=crime.rename({'Unnamed: 0':'State'}, axis=1)
```

```python
crime2
```

```python
#Normalization function
scaler = StandardScaler()
scaled_crime2 = scaler.fit_transform(crime2.iloc[:,1:])
```

```python
scaled_crime2
```

```python
#Using Elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_crime2)
    wcss.append(kmeans.inertia_) 
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#seen 4 clusters in above plot
clusters_new = KMeans(4, random_state=42) #randomly divide data into 4 partitions. Use any number but keep consistent
clusters_new
```

```python
clusters_new.fit(scaled_crime2)
clusters_new.labels_
```

```python
crime['ClusterID_new']=clusters_new.labels_
crime.head()
```

```python
#Average values of all data points within one cluster for profiling purposes.
clusters_new.cluster_centers_
```

```python
crime.groupby('ClusterID_new').agg(['mean']).reset_index()
```
